#### **Question 1: How do the number of installs vary across different app types and categories?**

````python
# Aggregate data: Count installs per app type and category
installs_grouped = merged_data.groupby(['Category', 'Type'])['Installs'].sum().reset_index()

pivot_installs = installs_grouped.pivot(index='Category', columns='Type', values='Installs')

pivot_installs.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Number of Installs by App Type and Category')
plt.xlabel('App Category')
plt.ylabel('Number of Installs')
plt.legend(title='App Type')
plt.xticks(rotation=50)
plt.show()
````

##### 1. Why did you pick the specific chart?


The chosen chart, a stacked bar chart depicting the number of installs by app type and category, is ideal for showcasing the distribution of installs across categories while emphasizing the contribution of each app type within these categories.
